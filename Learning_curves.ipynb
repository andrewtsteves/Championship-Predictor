```python
import matplotlib.pyplot as plt
import models_best_params as mbp
import validation as v
from sklearn.model_selection import learning_curve
import numpy as np
```

```python
estimators = [mbp.lr, mbp.svc, mbp.dt, mbp.rf]
labels = ['Logistic Regression', 'SVC', 'Decision Tree', 'Random Forest']
colors = ['red', 'aqua', 'green', 'yellow']
```

```python
fig, ax = plt.subplots(1, 2, figsize = (13, 6))

for estimator, label, color in zip(estimators, labels, colors):
    train_sizes_abs, train_scores, test_scores = learning_curve(estimator = estimator, X = mbp.X_train,
                                                 y = mbp.y_train, cv = v.rskf, scoring = 'accuracy',
                                                 train_sizes = np.linspace(0.2, 1.0, 10),
                                                 error_score = 0)

    train_mean = np.mean(train_scores, axis = 1)
    train_std = np.std(train_scores, axis = 1)

    test_mean = np.mean(test_scores, axis = 1)
    test_std = np.std(test_scores, axis = 1)

    ax[0].plot(train_sizes_abs, train_mean, label = label, color = color)
    ax[0].fill_between(train_sizes_abs, train_mean + train_std, train_mean - train_std, alpha = 0.1, color = color)

    ax[1].plot(train_sizes_abs, test_mean, label = label, color = color)
    ax[1].fill_between(train_sizes_abs, test_mean + test_std, test_mean - test_std, alpha = 0.1, color = color)

ax[0].set_xlabel('Number of training samples')
ax[0].set_ylabel('Training Score')
ax[1].set_xlabel('Number of test samples')
ax[1].set_ylabel('Test score')
plt.legend(loc= "lower right")
plt.suptitle('Scoring Type: Accuracy', fontsize = 20)
plt.show()
```
